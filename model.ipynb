```python
import pandas as pd
import numpy as np
from joblib import dump
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor ,AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
```

```python
df = pd.read_csv("Data.csv")
df.head()
```

```python
def years(date):
    year = str(date)
    yer = year[0:4]
    return int(yer)

# df["Date"].str.replace("-","")
df["year"] = df["Date"].apply(years)

df.drop("Date",axis=1,inplace=True)

df.head()
```

```python
df.dropna(axis=0,inplace=True)
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
col = df.drop(["IHSG","year"],axis=1).columns

scaler = StandardScaler()
df[col] = scaler.fit_transform(df[col])

df.head()
```

```python
X = df.drop('IHSG',axis=1)
y = df["IHSG"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
results = []
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Lasso Regression': Lasso(alpha=0.1, max_iter=10000),
    'ElasticNet': ElasticNet(alpha=0.1, l1_ratio=0.5),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42, n_jobs=-1),
    'Extra Trees': ExtraTreesRegressor(random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    'K-Neighbors': KNeighborsRegressor(),
    'Support Vector Regressor': SVR(kernel='rbf'),
    "XGBoost": XGBRegressor(random_state=42),
    "AdaBoost": AdaBoostRegressor(random_state=42),
}
```

```python
for name,model in models.items():
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    R2_score = r2_score(y_test,y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    MAE = mean_absolute_error(y_test, y_pred)
    RMSE = np.sqrt(MSE)
    results.append({
        "Name":name,
        "R2_score":R2_score,
        "MSE":MSE,
        "MAE":MAE,
        "RMSE":RMSE
    })

results_df = pd.DataFrame(results)
results_df
```

```python
model_final = GradientBoostingRegressor(random_state=42)
model_final.fit(X_train,y_train)
y_pred = model_final.predict(X_test)
r2_score(y_test,y_pred)
```

```python
import numpy as np
import pandas as pd
import joblib

print("--- ML Components Loading Test ---")
try:
    # 1. Artifacts ko load karna
    model = joblib.load('model1.pkl')
    scaler = joblib.load('scaler.pkl')
    columns = joblib.load('columns.pkl')
    print("✅ Success: Saari `.pkl` files perfectly load ho gayi hain!")
    print(f"Trained Columns List: {columns}\n")
    
    # 2. Test Input Data (Aap isme values badal kar check kar sakte hain)
    # Order: OIL, GOLD, USDIDR, SP500, VIX, CPI, BI_rate, US_rate
    test_year = 2026
    test_continuous_inputs = [[75.30, 1100.30, 9302.00, 1100.00, 20.00, 5.13, 6.50, 0.11]]
    
    print("--- Pipeline Execution Test ---")
    # 3. Continuous features ko scale karna (8 features)
    features_scaled = scaler.transform(test_continuous_inputs)
    
    # 4. Year ko aakhiri me stack karna (Jaise training me tha)
    year_input = np.array([[int(test_year)]])
    final_model_input = np.hstack((features_scaled, year_input))
    
    # 5. DataFrame banana training columns ke sath
    feature_names = ['OIL', 'GOLD', 'USDIDR', 'SP500', 'VIX', 'CPI', 'BI_rate', 'US_rate', 'year']
    final_df = pd.DataFrame(final_model_input, columns=feature_names)
    
    # 6. Prediction run karna
    prediction = model.predict(final_df)
    
    print(f"🚀 Prediction Result Successful!")
    print(f"Predicted IHSG Target Value: {prediction[0]:.3f}")

except Exception as e:
    print("❌ Error: Pipeline execution fail ho gayi!")
    print(f"Technical Error Details: {str(e)}")
```
